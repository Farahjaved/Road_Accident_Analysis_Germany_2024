```python
import geopandas as gpd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import folium
import fiona
```

```python
# Load the GeoPackage (contains multiple layers)
path = r"C:\Users\muham\Downloads\vg250_12-31.utm32s.gpkg.ebenen\vg250_12-31.utm32s.gpkg.ebenen\vg250_ebenen_1231\DE_VG250.gpkg"

print(fiona.listlayers(path))

```

```python
# Load the municipality layer
gdf = gpd.read_file(path, layer='vg250_gem')
gdf.head()
```

```python
# Keep only key fields
gdf = gdf[["GEN", "AGS", "geometry"]]

# Compute centroid coordinates for mapping
gdf["lon"] = gdf.geometry.centroid.x
gdf["lat"] = gdf.geometry.centroid.y

# Rename columns to match your clustering dataset
gdf = gdf.rename(columns={"GEN": "MuncipalityName"})
```

```python
# Import Data
# Define path
path = r'C:\Users\muham\OneDrive\Desktop\careerfoundry\Data Immersion\Achievement 6\05-10-25 Accident_data_germany_2024_analysis'
df_clustering_1_AGS = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data','df_clustering_1_AGS.csv'), low_memory= False)
```

```python
df_clustering_1_AGS.info()
```

```python
# Ensure AGS exists and is uniform
df_clustering_1_AGS['AGS'] = df_clustering_1_AGS['AGS'].astype(str).str.zfill(8)
gdf['AGS'] = gdf['AGS'].astype(str).str.zfill(8)
```

```python
# merging geo corrdinates
gdf_merged = gdf.merge(
    df_clustering_1_AGS,
    on='AGS',
    how='left'
)
```

```python
gdf_merged['cluster'].isna().mean()
```

```python
# Fixing encoding issues
df_clustering_1_AGS['MuncipalityName'] = (
    df_clustering_1_AGS['MuncipalityName']
    .str.encode('latin1', 'replace')
    .str.decode('utf-8', 'replace')
)

gdf['MuncipalityName'] = (
    gdf['MuncipalityName']
    .str.encode('latin1', 'replace')
    .str.decode('utf-8', 'replace')
)
```

```python
# reemerging
gdf_merged_new = gdf.merge(
    df_clustering_1_AGS,
    on='AGS',
    how='left'
)
```

```python
gdf_merged_new['cluster'].isna().mean()
```

```python
print(gdf.columns.tolist())
# If 'AGS' not present but 'RS' is, use:
if 'AGS' not in gdf.columns and 'RS' in gdf.columns:
    gdf = gdf.rename(columns={'RS':'AGS'})
```

```python
df_clustering_1_AGS['AGS'] = df_clustering_1_AGS['AGS'].astype(str).str.replace(r'\.0$', '', regex=True).str.zfill(8)
gdf['AGS'] = gdf['AGS'].astype(str).str.replace(r'\.0$', '', regex=True).str.zfill(8)

print(df_clustering_1_AGS['AGS'].str.len().value_counts())
print(gdf['AGS'].str.len().value_counts())
```

```python
chk = (
    df_clustering_1_AGS['state'].astype(str).str.zfill(2) +
    df_clustering_1_AGS['administrative_region'].astype(str).str.zfill(1) +
    df_clustering_1_AGS['administrative_district'].astype(str).str.zfill(2) +
    df_clustering_1_AGS['municipality'].astype(str).str.zfill(3)
)
print('same_as_existing_ags =', (chk == df_clustering_1_AGS['AGS']).mean())
```

```python
ags_c = set(df_clustering_1_AGS['AGS'])
ags_g = set(gdf['AGS'])

print('clustering AGS count:', len(ags_c))
print('geo AGS count:', len(ags_g))
print('intersection:', len(ags_c & ags_g))
print('coverage clustering->geo:', len(ags_c & ags_g)/len(ags_c))
print('coverage geo->clustering:', len(ags_c & ags_g)/len(ags_g))

# Show a few that don't match
unmatched_keys = list(ags_c - ags_g)[:20]
print('sample unmatched in geo:', unmatched_keys)
```
