## Data Cleaning


```python
#looking for missing values
df['OID_'].value_counts(dropna = False)
```

```python
for col in df.columns:
    types_in_col = df[col].map(type)
    if len(types_in_col.unique()) > 1:
        print(f"Column '{col}' has mixed data types: {types_in_col.unique()}")
```

```python
# Check for nulls in all columns
null_counts = df.isnull().sum()

# Display columns with at least one null value
print(null_counts[null_counts > 0])
```

```python
# finding full duplicates
df_dups = df[df.duplicated()]
```

```python
#changing column names to english
column_mapping = {
    "UIDENTSTLAE": "accident_id",
    "ULAND": "state",
    "UREGBEZ": "administrative_region", 
    "UKREIS": "administrative_district",
    "UGEMEINDE": "municipality",
    "UJAHR": "year_of_accident",
    "UMONAT": "month_of_accident",
    "USTUNDE": "hour_of_accident",
    "UWOCHENTAG": "day_of_week",
    "UKATEGORIE": "accident_category",
    "UART": "collision_type",
    "UTYP1": "accident_behavior_type",
    "ULICHTVERH": "light_conditions",
    "IstRad": "accident_with_bicycle",
    "IstPKW": "accident_with_pessenger_car",
    "IstFuss": "accident_with _pedestrian",
    "IstKrad": "accident_with_motorcycle",
    "IstGkfz": "involving_heavy_goods_vehicle",
    "IstSonstige": "other_vehicle_involved",
    "IstStrassenzustand": "road_surface_conditions",
    "LINREFX": "UTM_X", # X coordinate in ETRS89 / UTM Zone 32N (EPSG:25832)
    "LINREFY": "UTM_Y", # Y coordinate in ETRS89 / UTM Zone 32N
    "XGCSWGS84": "Longitude_WGS84", # Longitude in WGS84 (EPSG:4326), decimal degrees
    "YGCSWGS84": "Latitude_WGS84", # Latitude in WGS84 (EPSG:4326), decimal degrees
    "PLST": "location_check_level" # Level of success in geocoding the accident to a road section
}
df.rename(columns=column_mapping, inplace=True)
```
